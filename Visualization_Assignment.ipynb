**Q1 Plot Histograms using iris dataset**

```
plt.figure(figsize=(10,8))
#Histogram for sepal_length
plt.subplot(2, 2, 1)
plt.hist(iris["sepal_length"], bins=30, color='red')
plt.title("sepal_length")
plt.xlabel("sepal_length")
plt.ylabel("count");

#Histogram for sepal_width
plt.subplot(2, 2, 2)
plt.hist(iris["sepal_width"], bins=30, color="yellow")
plt.title("sepal_width")
plt.xlabel("sepal_width")
plt.ylabel("count");

#Histogram for petal_length
plt.subplot(2, 2, 3)
plt.hist(iris["petal_length"], bins=30, color="blue")
plt.title("petal_length")
plt.xlabel("petal_length")
plt.ylabel("count");

#Histogram for petal_width
plt.subplot(2, 2, 4)
plt.hist(iris["petal_width"], bins=30, color="green")
plt.title("petal_width")
plt.xlabel("petal_width")
plt.ylabel("count");

plt.tight_layout();
```

**Bivariate Analysis**

**Q13 Plot double Bar plot for the following data also add
legends**

```
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

fig, ax = plt.subplots(figsize=(10,5))
ax.bar(np.arange(len(labels)), men_means, color="blue", width=0.35, label='men_means')
ax.bar(np.arange(len(labels))+0.35, women_means, color="red", width=0.35, label='women_means')
## +0.4 -> mentioning the width postion from where he has to start draw

ax.set_xticks(np.arange(0.16,5, 1))  ##for positioning all labels 
ax.set_xticklabels(labels);

for x,g,s in zip(np.arange(5), men_means, women_means):
    ax.text(x-0.1, g+0.5, g) # annotating the men_means
    ax.text(x+0.3, s+0.5, s) # annotating the women_means

ax.set_title('Men and Women means', size=16)
ax.set_xlabel('labels', size=12) 
ax.set_ylabel('Number of means', size=12)
ax.legend(loc='best');
```

**Q14 Plot Pie Chart for the following data also add
legends**

```
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

fig, axe= plt.subplots(figsize= (8,5))
axe.pie(sizes, explode=explode, labels=labels, autopct='%1.0f%%', shadow= True, startangle= 90)   ## autopct -> represents the numers after decimal point
axe.axis('equal')      ##'equal' represents that figure should be a circle
axe.legend()
plt.show();
```
